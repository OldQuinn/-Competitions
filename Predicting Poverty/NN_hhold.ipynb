```python
def standardize(df, numeric_only=True):
    numeric = df.select_dtypes(include=['int64', 'float64'])
    
    # subtracy mean and divide by std
    df[numeric.columns] = (numeric - numeric.mean()) / numeric.std()
    
    return df
    

def pre_process_data(df, enforce_cols=None):
    print("Input shape:\t{}".format(df.shape))
        

    df = standardize(df)
    print("After standardization {}".format(df.shape))
        
    # create dummy variables for categoricals   使用get_dummies进行one-hot编码
    df = pd.get_dummies(df)
    print("After converting categoricals:\t{}".format(df.shape))
    

    # match test set and training set columns
    if enforce_cols is not None:
        to_drop = np.setdiff1d(df.columns, enforce_cols)
        to_add = np.setdiff1d(enforce_cols, df.columns)

        df.drop(to_drop, axis=1, inplace=True)
        df = df.assign(**{c: 0 for c in to_add})
    
    df.fillna(0, inplace=True)
    
    return df


 #used for B 因为b的数据不干净

from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
def split_data(data,fraction=0.2):
    data_y=data['poor']
#     lb=LabelBinarizer()
#     data_y=lb.fit_transform(data_y)
    data_x=data.drop(['poor'],axis=1)
    train_x,valid_x,train_y,valid_y=train_test_split(data_x,data_y,test_size=fraction)
    return train_x,valid_x,train_y,valid_y




def preprocessing(b_train_in,p=0.8):   
    # 1.null value processing
    ################################
    aa=[]
    bb=[]
    cc=[]
    for c in b_train_in.columns:
        if b_train_in[c].count()<10100:
            aa.append(c)
        elif b_train_in[c].count()==b_train_in.shape[0]:
            bb.append(c)
        else:
            cc.append(c)
    b_train_in.drop(aa,axis=1,inplace=True)
    for c in cc:
        b_train_in[c].fillna(b_train_in[c].mean(),inplace=True)


    
    #2. normalization
    #################################
    temp=b_train_in.select_dtypes(['int64','float64'])
    b_train_in[temp.columns]=(temp-temp.mean())/temp.std()
    
    #3. dummy encoder
    #################################
    b_train_in=pd.get_dummies(b_train_in)
    
    
    
    #4. VarianceThreshold
    #################################
    from sklearn.feature_selection import VarianceThreshold
    sel = VarianceThreshold(threshold=(p * (1 - p)))
    train=sel.fit_transform(b_train_in)
    
    
    
    #5. label encoder
    #################################
    from sklearn.preprocessing import LabelEncoder
    le=LabelEncoder()
    le.fit(['True','False'])
    b_train_in['poor']=le.transform(b_train_in['poor'])
    

    
#   6.split into features and values
#   #################################
    train_x,valid_x,train_y,valid_y=split_data(b_train_in)
    
    
    return train_x,valid_x,train_y,valid_y



```

```python
import keras
from keras.layers import Dense,Activation,BatchNormalization,Dropout
from keras import regularizers
from keras.optimizers import Adam
```

```python
model=keras.models.Sequential()
```

```python
model.add(Dense(200,input_shape=(859,)))   #,kernel_regularizer = regularizers.l2(0.005)
model.add(BatchNormalization())
model.add(Activation(activation='sigmoid'))
model.add(Dropout(0.7))

model.add(Dense(50))
model.add(BatchNormalization())
model.add(Activation(activation='sigmoid'))
model.add(Dropout(0.5))


model.add(Dense(10))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Activation(activation='relu'))


model.add(Dense(1))
model.add(BatchNormalization())
model.add(Activation(activation='sigmoid'))

model.compile(loss='binary_crossentropy',metrics=['acc'],optimizer=Adam(0.005))
model.summary()
```

```python
def make_country_sub(preds, test_feat, country):
    # make sure we code the country correctly
    country_codes = ['A', 'B', 'C']
    
    # get just the poor probabilities
    country_sub = pd.DataFrame(data=preds[:, 0],  # proba p=1
                               columns=['poor'], 
                               index=test_feat.index)

    
    # add the country code for joining later
    country_sub["country"] = country
    return country_sub[["country", "poor"]]
```
