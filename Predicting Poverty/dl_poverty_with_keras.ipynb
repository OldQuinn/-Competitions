1. 删除没用的列

```python
import numpy as np
y_trainC=trainC.poor
trainC1=trainC.drop(['id','iid'],axis=1)   #id是家庭编号，iid是每个家庭里面个人的编号
# trainC1.info()
```

```python
trainC1
```

```python
# 2. 标准化
```

```python
temp=trainC1.select_dtypes(['int64','float64'])
trainC1[temp.columns]=(temp-temp.mean())/temp.std()
```

 3. labelencoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(['True','False'])
trainC1['poor']=le.transform(trainC1['poor'])
```

```python
trainC1['poor']
```

4. dummy

```python
trainC1=pd.get_dummies(trainC1)   #这一步是将原来一个feature下面的分类变成新的feature，这样每个feature对应的都是0或1
```

```python
trainC1.head()
```

5. dataset split

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
def split_data(data,fraction=0.2): 
    data_y=data['poor']
#     lb=LabelBinarizer() 
#     data_y=lb.fit_transform(data_y) 
    data_x=data.drop(['poor'],axis=1)
    train_x,valid_x,train_y,valid_y=train_test_split(data_x,data_y,test_size=fraction)
    return train_x,valid_x,train_y,valid_y
```

```python
train_x_C,valid_x_C,train_y_C,valid_y_C=split_data(trainC1)
```

```python
print(train_x_C.shape)
print(valid_x_C.shape)
print(train_y_C.shape)
print(valid_y_C.shape)
```

```python
train_y_C = train_y_C.values.reshape(-1,1)
valid_y_C = valid_y_C.values.reshape(-1,1)
print(train_y_C.shape)
print(valid_y_C.shape)
```

6. model_build

```python
import keras
from keras.layers import Dense,Activation,BatchNormalization,Dropout
from keras.optimizers import Adam

```

```python
model=keras.models.Sequential()
bs=100
```

```python
model.add(Dense(50,input_shape=(300,)))
# model.add(BatchNormalization())
# model.add(Activation(activation='relu'))
# model.add(Dropout(0.2))
# model.add(Dense(20))
model.add(BatchNormalization())
model.add(Activation(activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.compile(loss='binary_crossentropy',metrics=['acc'],optimizer=Adam(0.001))
model.summary()
```

```python
model.fit(batch_size=100,epochs=200,validation_data=(valid_x_C,valid_y_C),x=train_x_C,y=train_y_C)
```
